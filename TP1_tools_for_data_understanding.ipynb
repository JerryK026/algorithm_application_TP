# 1.데이터를 이해하기 위한 도구들이 왜 필요한가?

우리가 실생활에서 이용하는 정보들은 자연에 있는 것들을 가져온 것들이다. 우리가 그러한 정보를 이용하려면 일련의 과정을 필요로 한다. 1.자연에 있는 무언가를 인식하고 (perception) 2.그것을 처리해서 (Science, Engeneering) 3.환경에 적용하는 것이다. (application) 즉, 우리가 데이터를 다루기 위해서는 우선 데이터를 먼저 파악할 수 있어야 한다. 이 때 사용하는 도구들은 대표적으로 확률, 통계, 선형대수가 있다.

# 2.일상생활에 어떻게 적용되는가?

a.통계 : 
통계라는 것은 사회집단에서의 상황을 숫자로 표현한 것이다. 즉, 통계라는 것은 과거에 있었던 일들 중, 우리가 원하는 값들(데이터)을 선별해낸 것이라고 할 수 있다. 우리는 이 통계를 기반으로 상황에 맞는 계산을 해서, 원하는 값을 유추해볼 수 있다. 예를 들어, 과자기업 A에서 10대의 입맛에 맞는 과자를 만든다고 했을 때, 10대가 선호하는 맛에 대한 통계를 기반으로 제품을 만들어 판매량을 더 늘릴 수 있다.

b.확률 : 
확률 확률은 전체 사건에서 특정한 사건의 수를 기반으로 나타낸 특정 사건이 일어날 기대수치이다. 확률은 항상 0보다 크고 1보다 작다. 이러한 확률은 Real World에서 데이터를 샘플링 후 Real World를 설명할 때 사용한다. 예를 들어, 복권은 당첨되기만 하면 엄청난 부를 손쉽게 얻을 수 있다. 그러나 많은 사람들은 복권을 사지 않는다. 복권에 당첨될 확률이 엄청나게 낮음을 알고 있기 때문이다.

c.선형대수 : 
데이터 처리에서 벡터가 중요한 이유는, 정보라는 것은 데이터들에서 원하는 것을 얻어내기 위해 어떠한 특징점들을 모아서 구체화시킨 것이다. 그런데, 벡터를 사용하면 데이터들을 수치화, 구체화할 수가 있다. 정보들은 공학, 생물학, 화학 등 일상생활에 거의 모든 분야에 이용할 수 있기 때문에 중요하다.

# 3.활용가능한 주요 알고리즘과 패키지 명

몬테카를로 알고리즘 : 난수를 이용해 물리적, 수학적 시스템의 행동을 시뮬레이션 하기 위한 계산 알고리즘. 통계학적이고, 비결정적이다. 적분 등으로 특정한 면적 구할 때, 프로젝트 관리 등에 사용된다.
 
라스베이거스 알고리즘 : 올바른 해를 빨리 출력하기 위해 확률적 선택을 한다. 항상 올바른 해를 출력한다. O(logn)의 평균속도를 가진다.

# 4.해당 알고리즘 혹은 주제의 핵심 아이디어

"정보"라는 것은 수많은 자료들 중 현재 필요로 하는 것들을 판별해내어 그것들을 의미 있게 규합한 것이다. 그리고 통계, 확률, 선형대수는 이렇게 만들어낸 정보들을 찾아내는 핵심 도구들이다. 우리는 이 도구를 통해서 얻어낸 정보들을 이용해서 알고리즘에 집어넣어 우리가 원하는 값들을 찾아내는 데 사용한다.
